# Data - Text Preprocessing
# 1. Lower Case
# 2. Tokenization
# 3. Removing Special Character
# 4 . Removing stop word and punctuation -> is of the
# 5 . Stemming / lamitization -> dance , dancing

```python
from nltk.corpus import stopwords
import string
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    y = []
    for i in text:
        if i.isalnum():   
            y.append(i) 
    text = y[:]
    y.clear()
    for i in text:
        if i not in stopwords.words("english") and i not in string.punctuation:
            y.append(i)
    text = y[:]
    y.clear()
    for i in text:
        ps = PorterStemmer()
        y.append(ps.stem(i))
       
        
        return " ".join(y)
```

```python
df["transform_text"] = df["text"].apply(transform_text)
```

# 4.Model Building

```python
df
print(df["transform_text"].isnull().sum())  # Check for missing values
df["transform_text"] = df["transform_text"].fillna("")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
```

```python
x = tf.fit_transform(df["transform_text"]).toarray()
```

```python
y = df["target"].values
```

```python
x ,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BernoulliNB
from sklearn.metrics import confusion_matrix , accuracy_score , precision_score
```

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
gnb.fit(x_train , y_train)
y_pred1 = gnb.predict(x_test)
score1 = accuracy_score(y_test , y_pred1)
score1
print(confusion_matrix(y_test , y_pred1))
print(precision_score(y_test , y_pred1))
```

```python
mnb.fit(x_train , y_train)
y_pred2 = mnb.predict(x_test)
print(accuracy_score(y_test , y_pred2))
print(confusion_matrix(y_test , y_pred2))
print(precision_score(y_test , y_pred2))
```

```python
bnb.fit(x_train , y_train)
y_pred3 = bnb.predict(x_test)
print(accuracy_score(y_test , y_pred3))
print(confusion_matrix(y_test , y_pred3))
print(precision_score(y_test , y_pred3))
```

```python
# Import necessary libraries
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.svm import SVC, LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier  # Requires XGBoost library

# List of classifiers
classifiers = {
    "Logistic Regression": LogisticRegression(),
    "SGD Classifier": SGDClassifier(),
    "Gaussian Naive Bayes": GaussianNB(),
    "Multinomial Naive Bayes": MultinomialNB(),
    "Bernoulli Naive Bayes": BernoulliNB(),
    "Linear SVM": SVC(kernel='linear'),
    "Non-linear SVM (RBF)": SVC(kernel='rbf'),
    "LinearSVC": LinearSVC(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Linear Discriminant Analysis": LinearDiscriminantAnalysis(),
    "Quadratic Discriminant Analysis": QuadraticDiscriminantAnalysis(),
    "Neural Network (MLP)": MLPClassifier(max_iter=1000),
    "XGBoost": XGBClassifier()  # Requires installation of xgboost
}
```

```python
# Import necessary metrics
from sklearn.metrics import accuracy_score, precision_score
import pandas as pd

# Initialize an empty list to store results
results = []

# Train and evaluate each classifier
for name, clf in classifiers.items():
    try:
        # Train the classifier
        clf.fit(x_train, y_train)

        # Make predictions
        y_pred = clf.predict(x_test)

        # Calculate metrics
        acc = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='binary')  # Use 'binary' for binary classification

        # Add results to the list
        results.append({
            "Model": name,
            "Accuracy": acc,
            "Precision": precision
        })

    except Exception as e:
        # Handle classifiers that fail
        print(f"{name} failed with error: {e}")

# Convert results to a DataFrame
results_df = pd.DataFrame(results)

# Sort the DataFrame by Precision
sorted_results_df = results_df.sort_values(by="Precision", ascending=False)

# Display the DataFrame
print(sorted_results_df)
```
